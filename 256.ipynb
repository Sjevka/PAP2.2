```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

Z=np.array([22,26,28,29,30,40,42,47])
#K_alpha (Ti, Fe, Ni, Cu, Zn, Zr, Mo, Ag) in keV:
K_alpha=np.array([4.46,6.42,7.51,8.08,8.68,15.81,17.89,21.95])
Delta_K_alpha=np.array([0.16,0.15,0.16,0.16,0.16,0.16,0.17,0.21])
sqrt_K_alpha=np.sqrt(K_alpha)
Delta_sqrt_K_alpha=0.5*Delta_K_alpha/sqrt_K_alpha

plt.errorbar(Z, sqrt_K_alpha, Delta_sqrt_K_alpha, fmt=".")
plt.xlabel('Kernladungszahl Z')
plt.ylabel(r'$\sqrt{E_\alpha}/\sqrt{keV}$ ', fontsize=14)
plt.title(r'$\sqrt{E_\alpha}$' + ' als Funktion von Z')
plt.savefig("figures/K_alpha_vs_Z_fit.pdf", format="pdf")

n1=1
n2=2
def fit_func(x, sqrt_Er, sig12):
    return sqrt_Er*(x-sig12)*np.sqrt(1/n1**2-1/n2**2)

from scipy.optimize import curve_fit
popt, pcov=curve_fit(fit_func, Z, sqrt_K_alpha,
sigma=Delta_sqrt_K_alpha)

plt.errorbar(Z, sqrt_K_alpha, Delta_sqrt_K_alpha, fmt=".")
plt.xlabel('Kernladungszahl Z')
plt.ylabel(r'$\sqrt{E_\alpha}/\sqrt{keV}$ ', fontsize=14)
plt.title(r'$\sqrt{E_\alpha}$' + ' als Funktion von Z')
plt.plot(Z, fit_func(Z,*popt))
plt.savefig("figures/K_alpha_vs_Z.pdf", format="pdf")

ChiQuad=sum(((popt[0]*(Z-popt[1])*np.sqrt(1/n1**2-1/n2**2)-sqrt_K_alpha)/Delta_sqrt_K_alpha)**2)

print("sqrt_Er=",popt[0], ",Standardfehler=",np.sqrt(pcov[0][0]))
print("sig12=",popt[1], ",Standardfehler=",np.sqrt(pcov[1][1]))
print("ChiQuad=",ChiQuad)
```
